#Find the Most Frequent Words in a String

```
def frequent_pattern(text, k):
  cnt, n = {}, len(text)
  for idx in range(n-k+1):
    key = text[idx:idx+k]
    cnt[key] = cnt.get(key, 0) + 1
  mx = max(cnt.values())
  return [key for key, val in cnt.items() if val == mx]
```

#Find Patterns Forming Clumps in a String

```
def clump_finder(genome, k, L, t):
  n, ans = len(genome) , set()
  mp = defaultdict(int)
  for idx in range(L-k+1):
    key = genome[idx:idx+k]
    mp[key] += 1
    if mp[key] >= t:
      ans.add(key)
  l , r = 0, L-k+1
  while r <= (n-k):
    key = genome[l:l+k]
    mp[key] -= 1
    key = genome[r:r+k]
    mp[key] += 1
    if mp[key] >= t:
      ans.add(key)
    l += 1
    r += 1
  return list(ans)
```

```
def hamming_distance(s1, s2):
  return sum(1 for idx, ch in enumerate(s1) if ch != s2[idx])
```

```
def frequent_pattern_with_mismatch(text, k, d):
  k_mers = [''.join(k_mer) for k_mer in product('ACGT', repeat=k)]
  n , mp = len(text) , defaultdict(int)
  for k_mer in k_mers:
    mp[k_mer] = sum(1 for idx in range(n-k+1) if hamming_distance(text[idx:idx+k], k_mer) <= d)
  for k_mer in k_mers:
    mp[k_mer] += sum(1 for idx in range(n-k+1) if hamming_distance(reverse_complement(text[idx:idx+k]),k_mer) <= d)
  mx = max(mp.values())
  return [key for key,val in mp.items() if val == mx]
```

```
def frequency_array(text, k):
  k_mers = [''.join(item) for item in product('ACGT', repeat=k)]
  mp , n = defaultdict(int) , len(text)
  for idx in range(n-k+1):
    mp[text[idx:idx+k]] += 1
  return [mp[k_mer] if k_mer in mp else 0 for k_mer in k_mers]
```

```
def pattern_to_number(pattern):
  mp = {'A':0, 'C':1, 'G':2, 'T':3}
  num = 0
  for ch in pattern:
    num = num * 4 + mp[ch]
  return num
```

```
def number_to_pattern(num,k):
  pattern = ''
  mp = {0:'A', 1:'C', 2:'G', 3:'T'}
  for _ in range(k):
    pattern += mp[(num)%4]
    num = num//4
  return pattern[::-1]
```
